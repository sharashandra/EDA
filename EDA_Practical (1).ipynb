```python
# Used for string(Non- Word Character) using string operations.
df['Name'] = df['Name'].str.replace(r'\W', '', regex = True)
```

```python
df['Domain'] = df['Domain'].str.replace(r'\W', '', regex = True)
```

```python
df['Age'] = df['Age'].str.replace(r'\W', '', regex = True)
print(df['Age'])
print(df)
```

```python
df['Age'] = df['Age'].str.extract('(\d+)')
df['Age']
```

```python
df['Location'] = df['Location'].str.replace(r'\W', '', regex = True)
df['Location']
```

```python
df['Salary'] = df['Salary'].str.replace(r'\W','', regex = True)
df['Salary']
```

```python
df
```

```python
df['Exp'] = df['Exp'].str.extract('(\d+)')
df['Exp']
```

```python
df
```

```python
clean_data = df.copy()
clean_data
```

# EDA TECHNIQUES

```python
clean_data.isnull().sum()
```

```python
clean_data['Age'] = clean_data['Age'].fillna(np.mean(pd.to_numeric(clean_data['Age'])))
clean_data['Age']
```

```python
clean_data['Exp'] = clean_data['Exp'].fillna(np.mean(pd.to_numeric(clean_data['Exp'])))
clean_data['Exp']
```

```python
clean_data['Location'] = clean_data['Location'].fillna(clean_data['Location'].mode()[0])
clean_data['Location']
```

```python
clean_data
```

```python
clean_data.info()
```

```python
clean_data['Age'] = clean_data['Age'].astype(int)
```

```python
clean_data['Salary'] = clean_data['Salary'].astype(int)
clean_data['Exp'] = clean_data['Exp'].astype(int)
```

```python
clean_data.info()
```

```python
clean_data['Name'] = clean_data['Name'].astype('category')
clean_data['Location'] = clean_data['Location'].astype('category')
clean_data['Domain'] = clean_data['Domain'].astype('category')
clean_data.info()
```
